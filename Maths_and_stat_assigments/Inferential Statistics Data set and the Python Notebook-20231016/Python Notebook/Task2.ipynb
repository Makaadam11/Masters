```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols


df = pd.DataFrame(pd.read_csv('../Dataset/PanicSymptoms.csv'))

df
```

```python
df1 = pd.DataFrame({'Cog': np.repeat([['Cog']], 30),
                   'Dosage': np.repeat([4, 8, 12], 10),
                   'Symptoms':[60, 60, 68, 67, 68, 67, 67, 70, 60, 61,
                               66, 65, 62, 58, 50, 59, 56, 60, 70, 61,
                               60, 67, 61, 66, 64, 61, 65, 68, 65, 67]})

df2 = pd.DataFrame({'Beh': np.repeat([['Beh']], 30),
                   'Dosage': np.repeat([4, 8, 12], 10),
                   'Symptoms':[66, 68, 67, 69, 68, 60, 66, 68, 62, 66,
                               56, 67, 65, 70, 62, 57, 60, 71, 65, 60,
                               59, 60, 65, 62, 57, 58, 57, 66, 61, 61]})

df3 = pd.DataFrame({'CogBeh': np.repeat([['CogBeh']], 30),
                   'Dosage': np.repeat([4, 8, 12], 10),
                   'Symptoms':[45, 45, 46, 42, 41, 46, 45, 45, 46, 44,
                               47, 46, 45, 45, 41, 42, 43, 40, 43, 46,
                               45, 45, 45, 42, 42, 43, 44, 45, 43, 43]})
df1
```

```python
model = ols('Symptoms ~ C(Cog) + C(Dosage) + C(Cog):C(Dosage)', data=df1).fit()
sm.stats.anova_lm(model, typ=1)

```

```python
model = ols('Symptoms ~ C(Beh) + C(Dosage) + C(Beh):C(Dosage)', data=df2).fit()
sm.stats.anova_lm(model, typ=1)

```

P < 0.5 so in Behavioral method we can acknowledge that drug dosage have significant impact on the outcome.

```python
model = ols('Symptoms ~ C(CogBeh) + C(Dosage) + C(CogBeh):C(Dosage)', data=df3).fit()
sm.stats.anova_lm(model, typ=1)
```
