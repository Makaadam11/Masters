Numeric Types:int, float, complex

Mapping Type:dict
Set Types:set, frozenset
Boolean Type:bool
Binary Types:bytes, bytearray, memoryview

```python
# Multiple Columns in Y axis
# importing pandas library 
import pandas as pd 
# import matplotlib library 
import matplotlib.pyplot as plt 

# creating dataframe 
df = pd.DataFrame({ 'Name': ['Paul', 'Adam', 'Grace'], 'Age': [20, 32, 65], 'Baggage Allowance (kgs)': [22, 44, 66] }) 

# plotting graph 
df.plot(x="Name", y=["Age", "Baggage Allowance (kgs)"], kind="bar") 

```

```python
# Same Axis Plot
# importing pandas library 
import pandas as pd 
# import matplotlib library 
import matplotlib.pyplot as plt 

# creating dataframe 
df = pd.DataFrame({ 'Name': ['Paul', 'Adam', 'Grace'], 'Age': [20, 32, 65], 'Baggage Allowance (kgs)': [22, 44, 66] }) 

# plotting Baggage Allowance (kgs)
ax = df.plot(x="Name", y="Baggage Allowance (kgs)", kind="bar") 
# plotting age on the same axis 
df.plot(x="Name", y="Age", kind="bar", ax=ax, color="maroon") 
```
