# Visualization #

## Bar Plot/Chart ##
Mostly used to visualize complex and confusing data and It is mostly used to compare measures (like mean, sum, frequency, amount, etc) for distinct categories of data by varying the heights or lengths. 
**When to choose Bar charts:** In short to plot "Measured value" for "Varied/distict catagories of data"
**Types of Bar charts:** Vertical, Horizontal, Grouped, Stacked
**Preferred Use case:** If you have categorical data or ordinal data, a bar plot is an appropriate choice. 

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

Degree = ['BScDSAI', 'BSCCS', 'MScDSAI', 'MSCCS']
counts = [15, 20, 25, 30]
bar_labels = ['red','_red','blue','_blue']
bar_colors = ['tab:red','tab:red','tab:blue','tab:blue']

ax.bar(Degree, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Academic Admissions at CCI')
ax.set_title('Admission count by Degree and UG/PG Grouping')

plt.show()
```

# Scatter Plot #
A scatter plot is a type of data visualization that shows the relationship between different variables or in other words to check if to determine whether or not two variables have a relationship or correlation. 
Used to measure correlation but not a law of causation. 
**When to choose Scatter Plot:** 1. When you are interested to determine relation between Independent and Dependent variables 2.When your independent variable has multiple values for your dependent variable 3. When you have two variables that pair well together
**When not to use Scatter Plot:**: 1. when your data is not at all related 2. when you have too large a set of data.
**Signs of Correlation:**: 

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)


fig, ax = plt.subplots()
for color in ['blue', 'orange', 'green']:
    n = 100
    x, y = np.random.rand(2, n)
    scale = 200 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color,
               alpha=0.7, edgecolors='none')

ax.legend()
ax.grid(True)

plt.show()
```
